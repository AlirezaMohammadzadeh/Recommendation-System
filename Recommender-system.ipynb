# Collaborative filtering

This method makes recommendations based on the past behavior of similar users. For example, if several users who are similar to a particular user have bought a certain product, that product might be recommended to the target user.

# Content-based filtering

This method makes recommendations based on the attributes of the products that a user has liked in the past. For example, if a user has liked several products that are red and made of silk, a content-based recommender might recommend other red silk products to that user.

## Two main types of collabrative filtering

### User-based collaborative filtering

This method finds other users who are similar to the target user based on their past behavior (e.g. which products they have bought or rated), and then makes recommendations based on the products that those similar users liked. For example, if several users who are similar to the target user have bought a certain product, that product would be recommended to the target user

### Item-based collaborative filtering

Instead of finding similar users, this method finds items that are similar to the items that the target user has liked in the past, and then recommends those similar items. It looks at the items that similar users have interacted with, and recommend items that are often co-purchased or co-rated together.

## Collaborative filtering Implementation

Both methods can be implemented by creating a similarity matrix that quantifies the similarity between users or items, which can be calculated using techniques such as cosine similarity or Pearson correlation. These matrices are then used to generate recommendations for a target user.

Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them

In the context of a recommendation system, cosine similarity is often used to measure the similarity between users or items in a user-item matrix. It is a way of quantifying how similar two users or two items are based on the products they've rated or bought.

### The value of cosine

The value of cosine similarity ranges from -1 to 1. If the value is 1, it means the vectors are identical, and if the value is -1, it means the vectors are completely dissimilar. The closer the value is to 1, the more similar the vectors are.

# Content based filtering

Content-based filtering is a technique for making recommendations based on the characteristics or attributes of the items themselves, rather than on the past behavior of users. It is used to recommend items that are similar to those that a user has previously interacted with or expressed interest in.

## Basic Idea

The basic idea behind content-based filtering is to represent each item as a set of features or attributes, and then use those features to measure the similarity between items. For example, if you were building a movie recommendation system, you might use features such as the actors, director, genre, and plot summary to represent each movie

## Main steps

### Step 1

Representing items as feature vectors: Each item is represented as a vector of its features. For example, a movie could be represented as a vector of its genre, director, and actors. The length of the vector depends on the number of features you choose to represent the items with.

### Step 2

Measuring similarity between items: Once the items have been represented as feature vectors, a similarity measure is used to compare the feature vectors of the items. Common similarity measures include cosine similarity, Euclidean distance, and Jaccard similarity.

## Euclidean distance

Euclidean distance is a measure of the distance between two points in a multidimensional space. It is a commonly used similarity measure in content-based filtering for recommendation systems.

In a content-based filtering system, each item is represented as a vector of its features. The Euclidean distance between two vectors is calculated as the square root of the sum of the squared differences between the corresponding elements of the vectors.

The Euclidean distance between two vectors is always non-negative, and it is equal to zero if and only if the vectors are identical. As the distance between the vectors increases, the similarity decreases

It is a simple and easy to understand measure, but it doesn't work well with sparse data, or when there are missing values, because it is sensitive to the scale of the features. Additionally, Euclidean distance can be sensitive to outliers, meaning that one very different feature can greatly increase the distance and make two items look dissimilar, even if they are quite similar in other features.

In a recommendation system, the smaller the Euclidean distance, the more similar the items are.

It is commonly used in content-based filtering to measure the similarity between items in a recommendation system, but it's not the only one and you might have to test different similarity measures to find the one that works best for your specific use case and dataset.

## Jaccard similarity

Jaccard similarity is a measure of similarity between two sets, it is based on the ratio of the size of the intersection of the sets to the size of the union of the sets.

In the context of a recommendation system, Jaccard similarity is commonly used to measure the similarity between items in a content-based filtering system, particularly when the items are represented as binary sets of features.

The Jaccard similarity coefficient ranges from 0 to 1, with 1 indicating that the sets are identical and 0 indicating that the sets have no elements in common.

It is a simple and effective measure of similarity when items are represented as binary sets of features. It is particularly useful when working with sparse data and can handle missing values well. However, it can be sensitive to the scale of the features, meaning that if one item has a lot more features than another, it can be considered more similar even if the features they share are less relevant.

In a recommendation system, the bigger the Jaccard Similarity, the more similar the items are.

### Encoding

Encoding categorical variables, such as brand, is an important step in preparing the data for a content-based filtering system. There are several ways to encode categorical variables, each with its own advantages and disadvantages. Here are a few common methods:

One-hot encoding: This method creates a new binary feature for each unique value of the categorical variable. For example, if there are three brands, A, B, and C, three new binary features would be created, one for A, one for B, and one for C. Each item would then be represented by a vector of ones and zeroes, with a one in the position corresponding to the brand of the item. One-hot encoding is simple and easy to understand, but it can create a high-dimensional feature space, which can be computationally expensive.

Ordinal encoding: This method assigns an integer value to each unique value of the categorical variable, based on a predefined ordering. For example, A=1, B=2, C=3. This method requires that the values have an inherent order, which might not always be the case, and can introduce bias if the order is not accurate.

Binary encoding: This method encodes categorical variables as a binary code. This is a more efficient way than one-hot encoding when you have a large number of categories, it reduces the dimensionality of the feature space.

Embeddings: This method encodes categorical variables as low-dimensional continuous vectors, which can be used as input to neural networks. This method is useful when working with large numbers of categories and can capture complex relationships between categories.

#### Embedding

An embedding is a low-dimensional, continuous vector representation of a discrete variable, such as a word or a brand. Embeddings are typically learned from data using neural networks, and they are used to capture the relationships between the discrete variables in a way that can be used by machine learning models.

In the context of content-based filtering, embeddings can be used to represent the categorical variables, such as brand, in a low-dimensional space, which allows the model to capture the relationships between the items in a more accurate and efficient way.

In the case of the shopping website content-based filtering, the model learns to represent the items in a continuous, low-dimensional space by adjusting the weights of the embedding layer during the training process.

The model is trained on a dataset that contains the items and their features, such as brand, color, price, etc. The input to the model is a batch of items and their encoded features, and the output is the same items encoded in the low-dimensional space defined by the embedding layer.

The model is trained using an optimization algorithm such as Adam, which adjusts the weights of the embedding layer in a way that minimizes the difference between the input and output items.

In this case, the model learns the relationship between the items and their features, and it encodes these relationships in the form of the embeddings. Once the model is trained, the embeddings can be used to find the similarity between the items using a similarity measure such as cosine similarity, in order to make recommendations to the users.
